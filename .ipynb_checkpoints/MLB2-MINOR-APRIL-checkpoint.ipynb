```python
#importing libraries:

#for basic mathematical and dataframe operations:

import numpy as np 
import pandas as pd 
from math import * 

# for visualizations:

import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#describing the data:

data.describe()
```

```python
#changing the spaces to underscore:

data.columns = data.columns.str.strip().str.lower().str.replace(' ','_').str.replace('/','_')
```

```python
#checking if the data contains any NULL values:

data.isnull().sum()
```

```python
#Providing a passingmark criteria which will be used to categorize the students:

passmarks = 40 
```

```python
#Making some few new features:

#Total marks:

data['total_score'] = data['math_score'] + data['reading_score'] + data['writing_score']

#Result:

data['result'] = np.where(data['total_score']<120, 'F', 'P')

# creating a new column pass_math, this column will tell us whether the students are pass or fail
data['pass_math'] = np.where(data['math_score']< passmarks, 'Fail', 'Pass')
data['pass_reading'] = np.where(data['reading_score']< passmarks, 'Fail', 'Pass')
data['pass_writing'] = np.where(data['writing_score']< passmarks, 'Fail', 'Pass')

#Percentage:
data['percentage'] = data['total_score']/3


data.head()
```

```python
# checking which student is fail overall:

data['status'] = data.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or 
                           x['pass_reading'] == 'Fail' or x['pass_writing'] == 'Fail'
                           else 'pass', axis = 1)

plt.rcParams['figure.figsize'] = (15, 8)
sns.countplot(data['status'], palette = 'muted')
plt.title('overall results', fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
# Assigning grades to the grades according to the following criteria :
# 0  - 40 marks : grade E
# 41 - 60 marks : grade D
# 60 - 70 marks : grade C
# 70 - 80 marks : grade B
# 80 - 90 marks : grade A
# 90 - 100 marks : grade O


def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

data['grades'] = data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

# plotting a pie chart for the distribution of various grades among the students:

labels = ['Grade O', 'Grade A', 'Grade B', 'Grade C', 'Grade D', 'Grade E']
sizes = data['grades'].value_counts()
colors = plt.cm.BuPu(np.linspace(0, 1, 7))
explode = (0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001)

plt.pie(sizes, colors = colors, labels = labels, shadow = True, autopct = '%.2f%%', startangle = 90)
plt.legend(loc = 'upper right')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
